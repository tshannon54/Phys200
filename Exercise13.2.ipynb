```python
from string import punctuation, whitespace, digits  def breakfile(filename):        d= dict()        myfile = open(filename)        for line in myfile:            process_line(line, d)        return ddef process_line(line, d):          line = line.replace('-', ' ')        for word in line.split():                word = word.strip(punctuation + whitespace + digits)                word = word.lower()                if word != '': #I had issues, so I had to use some of his solution here.                        d[word] = d.get(word, 0) + 1                 book1= breakfile('theGirlWhoHadNothingBook.txt') book2 =breakfile('theMarriageofEstherBook.txt')
```

```python
def sortedDictValues1(adict):   #this program sorts all the words in the list in alphabetical order    items = adict.items()    items.sort()    return items
```

```python
def sortedDictValues2(adict):   #this program sorts all the words in the list by number of times they appeared in the book    items = adict.items()    backitems=[ [v[1],v[0]] for v in items]    backitems.sort()    return backitems
```

```python
sortedDictValues2(book1)#the word "the" was used 2001 times in "The Girl Who Had Nothing"
```

```python
sortedDictValues2(book2)#the word "the" was used 3245 times in "The Marriage of Esther"
```
