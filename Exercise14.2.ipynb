```python
import sysdef sed(patternstr, replstr, file1, file2):    try:        fin = open(file1, 'r')        fout = open(file2, 'w')        for line in fin:            line = line.replace(patternstr, replstr)            fout.write(line)        fin.close()        fout.close()    except:        print 'Something went wrong.'    
```
